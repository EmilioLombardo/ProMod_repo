# Plotting av lufttrykk mot høyde
Ved hjelp av Appen phyphox og barometeret i telefonen, målte vi lufttrykket inne på skolen ved forskjellige høyder.

### Fremgangsmåte for måling
Vi gjorde målinger ved jevne høydeintervaller (6 trappetrinn) fra gulvninvået i nederste etasje til gulvnivået i øverste etasje. (Det er 24 trappetrinn per etasje med unntak av den nederste, som har kun 18 trinn.) For hvert høydeintervall målte vi lufttrykket i ca ett minutt. Dessverre fikk vi ikke overført alle målingene vi gjorde, så datasettet brukt her er dessverre litt lite :/

### Datahåndtering og plotting i python
For å få ett tall på trykket for hver høyde tok vi gjennomsnittet av lufttrykket målt over ett minutt. Så plottet vi lufttrykk mot høyde (målt i antall trappetrinn fra bunnen).

```python
from pylab import *

trinn = []
trykk = []

# Les csv-filer og lag dataliste
for i in range(8):
    trinn.append(i*6)
    data = loadtxt("trykkmålinger/" + str(i*6) + ".csv", skiprows=1)
    snitt = sum(data[:,1])/len(data[:,1])
    trykk.append(snitt)

# Gjennomsnitt og standardavvik
gj_snitt = mean(trykk)
avvik = std(trykk)

print(f"Gjennomsnittstrykk: {round(gj_snitt, 3)}\n"
    + f"Standardavvik: {round(avvik, 3)}")

# Graf
plot(trinn, trykk, marker=".")
xlabel("Høyde i trappetrinn")
ylabel("Trykk i hPa")

# regresjon
a,b = polyfit(trinn, trykk, 1)
x_range = array(range(0,46))
plot(x_range ,a*x_range + b)

show()
```

### Analyse
Gjennomsnittet på 996,03 hPa forteller at det er normaltrykk inne på skolen, og det lave standardavviket forteller at variasjonen i trykk for de første par etasjene er lav.

Regresjonen viser at trykket synker med høyden. Ut i fra regresjonen kan man anta at trykket vil fortsette å synke jo høyere oppe man måler.
