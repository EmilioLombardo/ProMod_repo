# Jordas temperatur — Modelleringsprosjekt

### Hensikt

Vi skal konstruere noen enkle modeller for jordas temperatur, teste dem, sammenligne resultatene med virkeligheten, og se hva modellene kan fortelle oss om noen av faktorene som påvirker jordas temperatur.

---

# Teori

Det fins utallige måter å lage en modell for jordas temperatur på. Det er såpass mange faktorer og systemer som spiller inn at det er vanskelig å skape nøyaktige modeller, selv for et kort tidsspenn. Likevel er modeller for jordas temperatur svært nyttige i tilknytning til forskning på klima og global oppvarming. Man kan for eksempel anvende temperaturmodeller for å prøve å forutsi konsekvenser av forskjellige tiltak mot global oppvarming (eller evt. mangelen på dem).

Vi skal her begynne ganske forenklet, og etter hvert bygge ut modellene for å ta hensyn til flere faktorer og anvendelser.

---

### Modell 1

For modell 1 skal vi lage en enkel modell hvor vi kun tar hensyn til solas stråling på jorda, uten atmosfære — uten drivhuseffekt.

##### Beregne solas utstråling

Vi må finne hvor mye energi sola utsråler, og så hvor stor del av denne strålingen som treffer jorda.

For å finne hvor mye energi sola utsråler kan man bruke _Stefan-Boltzmanns lov,_ som forteller hvor mye varmestråling som kommer fra et legeme (i $W/m^{2}$) gitt temperaturen til legemet (i Kelvin). 

$$S = \sigma T^{4}$$

Her er $\sigma$ Stefan-Boltzmanns konstant; $5,6703 \cdot 10 ^ {−8} \frac{W}{m^{2}K^{4}}$

Setter man inn solas temperatur for $T$, kan man beregne varmeutstrålingen per kvadratmeter fra sola $S_{sol}$.

##### Beregne solinnstråling på jorda

Jo lenger unna sola man er, jo større areal er solstrålingen fordelt på; den blir svakere jo lenger unna sola man er. Altså må vi regne ut akkurat hvor mye svakere strålingsstyrken blir _ved jorda._ Da må vi se på overflaten til en imaginær kule med sentrum i sola og radius lik avstanden mellom jorda og sola. På denne avstanden fra sola vil strålingen måtte fordeles på overflaten til den imaginære kulen.

Strålingsstyrken ved jorda $S_{0}$ er da solstyrken ved sola $S_{sol}$ ganget med forholdet mellom overflaten til sola og overflaten til den imaginære kulen:

$$
S_{0} = \frac{4 \cdot \pi \cdot radius_{sol}^{2}}{4 \cdot \pi \cdot D^{2}} \cdot S_{sol}
$$

<center>($D$ er avstand mellom jorda og sola)</center>

Dette kan forenkles:
$$
S_{0} = \frac{radius_{sol}^{2}}{D^{2}} \cdot S_{sol}
$$

##### Beregne gjennomsnittstemperatur på jorda

Vi gjør oss den antakelsen at temperaturen til jorda er konstant. Det impliserer at strålingen som jorda absorberer er like mye som den sender ut i varmestråling; $E_{inn} = E_{ut}$.

$E_{inn}$ er ikke helt lik $S_{0}$, fordi en viss del vil reflekteres (gitt ved _albedo_ — refleksjonsfaktoren). Dessuten må vi gange opp $S_{0}$ med arealet som treffes, siden $S_{0}$ er gitt i watt _per kvadratmeter_. Dette arealet er likt arealet til en _flat disk_ med jordas radius, siden jorda fra solas perspektiv ser ut som en sirkel:



$E_{ut}$ er jordas varmestråling. Strålingsstyrken, watt _per kvadratmeter,_ er gitt ved Stefan-Boltzmanns lov. Strålingen sendes ut fra hele jordas overflate, så vi må gange med hele jordas overflate, altså *kule*overflate, for å få den totale strålingseffekten.

$$E_{inn} = E_{ut}$$

$$S_{0}\cdot (1-albedo)\cdot (\pi r_{jord}^{2}) = \sigma T_{jord}^{4}\cdot (4\pi r_{jord}^{2})$$

Og nå gjenstår bare å løse for jordas gjennomsnittstemperatur $T_{jord}$.
  
$$\frac{S_{0}\cdot (1-albedo)}{4} = \sigma T_{jord}^{4}$$  

<center><span style="color:gray">($\pi r^{2}$ strykes)</span></center>
  
$$T_{jord} = \sqrt[4]{\frac{S_{0}\cdot (1-albedo)}{4\sigma}}$$

### Modell 2

I denne modellen skal vi inkludere en forenklet atmosfære for å prøve å få med drivhuseffektens innvirkning på jordas temperatur.

Vi gjør oss her noen antagelser:
* All solinnstråling går rett til jorda uten at noe absorberes av atmosfæren
* Atmosfæren absorberer 100% av jordas varmestråling

(Temperatur kan sees på som energi per areal (J/m<sup>2</sup>), så temperaturendring kan sees på som energi per areal per sekund, eller effekt per areal (W/m<sup>2</sup>). Slik går enhetene opp i de følgende uttrykkene.)

##### Temperaturendring i atmosfæren

Temperaturendring til atmosfæren er altså gitt ved varmestråling inn fra jorda minus atmosfærens egen varmestråling. Det gir følgende diff.likning:

$$T'_{atm} = \sigma T_{jord}^{4} - \sigma T_{atm}^{4}$$


##### Temperaturendring på jorda

Noen variabler/konstanter for å gjøre uttrykkene litt mer lesbare:
* Total solstråling inn til jorda (watt): $P_{inn} = S_0 \cdot (1 - albedo) \cdot \pi r_{jord}^{2}$
* Jordas overflate (m<sup>2</sup>): $A = 4 \pi r_{jord}^{2}$
    * Vi antar her at atmosfæren er en kule med samme overflate som jorda
* Hvor stor andel av atmosfærens varmestråling som kommer tilbake til jorda: $\alpha$
    * Vi kan for denne modellen anta at atmosfæren emitterer like mye stråling til jorda som ut mot verdensrommet. Da blir $\alpha$ = 0,5.

Jordas energiendring (watt) er da gitt ved følgende diff.likning:

$$
E'_{jord} = P_{inn} + \alpha \cdot \sigma T_{atm}^{4} \cdot A - \sigma T_{jord}^{4} \cdot A
$$

Og så kan vi dele på areal for å få temperaturendring (W/m<sup>2</sup>):

$$
T'_{jord} = \frac{P_{inn}}{A} + \alpha \cdot \sigma T_{atm}^{4} - \sigma T_{jord}^{4}
$$

For å få temperaturen til jorda, må man løse diff.likningen, for eksempel med Eulers metode.

---

### Modell 3

Her skal vi modifisere og utvide modell 2 for å (forhåpentligvis) passe bedre med virkeligheten. Vi antar ikke lenger at solinnstrålingen ikke absorberes av atmosfæren, eller at $\alpha$ = 0,5. I tillegg tar vi hensyn til at noe av jordas varmestråling ikke absorberes av atmosfæren.

Da introduseres nye koeffisienter. I tillegg til $\alpha$ har vi nå også $\beta$ og $\gamma$:
* $\alpha$: Hvor stor del av atmosfærens varmestråling som sendes mot jorda
* $\beta$: Hvor stor del av jordas varmestråling som absorberes av atmosfæren
* $\gamma$: Hvor stor del av solinnstrålingen som absorberes av atmosfæren

Diff.likningen for temperaturen til atmosfæren blir da:
$$
T'_{atm} = \gamma \cdot S_0 \cdot (1-albedo) + \beta \cdot \sigma T_{jord}^{4} - \sigma T_{atm}^{4}
$$

Og diff.likningen for jordas temperatur:
$$
T'_{jord} = \frac{(1 - \gamma) \cdot S_0 \cdot (1-albedo)}{A} + \alpha \cdot \sigma T_{atm}^{4} - \sigma T_{jord}^{4}
$$

For å finne rimelige verdier for $\alpha$, $\beta$ og $\gamma$ bruker vi den følgende oversikten for jordas energibudsjett fra USAs National Weather Service.


https://www.weather.gov/jetstream/energy

Herfra kan vi tolke at
* $\alpha = 98/(98 + 49 + 9) = 0,6282$
* $\beta = 104/116 = 0,8966$
* $\gamma = (19 + 4)/(19 + 4 + 47) = 0,3286$

Disse tallene underbygges også av denne siden om jordas energibudsjett fra Nasa (2014): https://science-edu.larc.nasa.gov/wp-content/uploads/sites/141/2020/07/ERB-poster-combined-update-8.2019v4.pdf

### Modell 4

Her skal vi utvide modell 2 og forsøke å modellere temperaturen til ett punkt på jorda gjennom et helt år.

##### Jordas varierende avstand til sola

For å modellere temperatur på jorda i løpet av et år, må vi ta hensyn til den varierende avstandten mellom jorda og sola. Jordas bane rundt sola er en oval, så avstanden til sola er ikke konstant. Den minste avstanden er ~146,000,000 km, og den maksimale avstanden er ~152,000,000 km. Vi kan modellere denne avstanden som funksjon av tid gitt i dager etter 4 januar, som er den dagen jorda er nærmest sola. En tilnærming vi kan bruke er en cosinusfunksjon med periode på 365 dager og som svinger mellom maksimum og minimum avstand mellom jorda og sola.

En generell cosinusfunksjon som starter ved bunnpunkt:
$$
D(t) = -A cos(\frac{2\pi}{P}t) + d
$$
* $A$ er amplituden, gitt ved $\frac{D_{max} - D_{min}}{2}$
* $d$ er likevektslinja, gitt ved $\frac{D_{max} + D_{min}}{2}$
* $P$ er perioden

Hvis vi fyller inn med $D_{max} = 1,52 \cdot 10^{11}$, $D_{min} = 1,46 \cdot 10^{11}$ (meter), og $P = 365$ (dager), får vi:

$$
D(t) = -3 \cdot 10^{9} cos(\frac{2\pi}{365}t) + 1,49 \cdot 10^{11}
$$

##### Hvilket punkt på jorda?



Vi velger et punkt på den nordlige halvkule, for eksempel ved norge. Solinnstrålingen vil være svakere her; strålingen vil være fordelt på et større areal på grunn av at overflaten ved Norge _ikke_ står vinkelrett på solstrålingsretningen slik som overflaten ved ekvator ville vært, og solstrålingens effekt per kvadratmeter ($S_0$) vil da bli lavere.

Vi ser også bort fra jordas helning. Da trenger vi kun å variere distansen mellom jorda og sola for hver dag, og regne ut temperaturen på samme måte som i modell 2 for alle dagene i året.

---

# Resultater

Her skal vi se på alle modellene i python, og finne hvilke resultater de gir for jordas temperatur.

### Modell 1

```python
from pylab import *

# Konstanter

sigma = 5.6703e-8      # Stefan-Boltzmanns konstant
T_sol = 5778           # Temperatur til sola (kelvin)
r_sol = 696_340_000    # Radius til sola (meter)
r_jord = 6_371e3       # Radius til jorda (meter)
D = 149_597_870_000    # Avstand mellom sola og jorda (meter)
albedo = 0.3           # Refleksjonsfaktor for jorda

# Beregninger

S_sol = sigma * T_sol**4 # Solas strålingsstyrke ved soloverflaten

S_0 = S_sol * r_sol**2/D**2 # Solas strålingsstyrke ved jorda

T_jord = (S_0 * (1 - albedo)/(4 * sigma))**(1/4) # Snittemperatur på jorda (i kelvin)

print(f"Solas varmestråling per kvadratmeter: {S_sol:e} W/m²")
print(f"Solstrålingens styrke ved jordas atmosfære: {round(S_0, 2)} W/m²")
print("\n" + f"Gjennomsnittlig temperatur på jorda: {round(T_jord - 273.15)}˚C")
```

### Modell 2

```python
from pylab import *

# --- Konstanter ---

sigma = 5.6703e-8       # Stefan-Boltzmanns konstant
T_sol = 5778            # Temperatur til sola (kelvin)
r_sol = 696_340_000     # Radius til sola (meter)
r_jord = 6_371e3        # Radius til jorda (meter)
D = 149_597_870_000     # Avstand mellom sola og jorda (meter)
albedo = 0.3            # Refleksjonsfaktor for jorda
A = 4 * pi * r_jord**2   # Jordas kuleoverflate (m^2)
                                                                                
alfa = 0.5  # Andel av atmosfærens stråling som kommer tilbake til jorda.

S_sol = sigma * T_sol**4    # Solas strålingsstyrke ved soloverflaten (W/m^2)
S_0 = S_sol * r_sol**2/D**2 # Solas strålingsstyrke ved jorda (W/m^2)

P_inn = S_0 * (1 - albedo) * (pi * r_jord**2)# Solstrålingens effekt på jorda (W)

# --- Diff.likninger for temperaturendring ---

# Temperaturendring i atmosfæren
def T_a_der(T_j, T_a):
    return (sigma * T_j**4) - (sigma * T_a**4)

# Temperaturendring i atmosfæren
def T_j_der(T_j, T_a):
    return (P_inn/A) + alfa * (sigma * T_a**4) - (sigma * T_j**4)


# --- Løsning med Eulers metode ---

def eulers_metode(T_j, T_a, T_j_der, T_a_der, utdyp=False, returner=False):

    T_j_lst = []
    T_a_lst = []

    t_lst = []
    t_maks = 8
    dt = 1e-3

    # Initialbetingelser
    T_j_0 = T_j
    T_a_0 = T_a

    for t in arange(0, t_maks, dt):
        T_j = T_j + T_j_der(T_j, T_a) * dt
        T_a = T_a + T_a_der(T_j, T_a) * dt

        t_lst.append(t)
        T_j_lst.append(T_j)
        T_a_lst.append(T_a)

    if utdyp:
        # Plotting og printing
        plot(t_lst, T_j_lst, label="Jordas temperatur")
        plot(t_lst, T_a_lst, label="Atmosfærens temperatur")

        xlabel("Tid")
        ylabel("Temperatur i kelvin")
        legend()
        show()

        print(f"Initialbetingelser: T_j(0) = {T_j_0}˚C, T_a(0) = {T_a_0}˚C")
        print(f"Jordas temperatur stabiliserer seg rundt {round(T_j_lst[-1] - 273.15, 2)}˚C")
        print(f"Atmosfærens temperatur stabiliserer seg rundt {round(T_a_lst[-1] - 273.15, 2)}˚C")
    
    if returner:
        return T_j, T_a


# --- Finne T_j og T_a ---

eulers_metode(0, 0, T_j_der, T_a_der, utdyp=True)
```

### Modell 3

```python
from pylab import *

# --- Konstanter ---

sigma = 5.6703e-8       # Stefan-Boltzmanns konstant
T_sol = 5778            # Temperatur til sola (kelvin)
r_sol = 696_340_000     # Radius til sola (meter)
r_jord = 6_371e3        # Radius til jorda (meter)
D = 149_597_870_000     # Avstand mellom sola og jorda (meter)
albedo = 0.3            # Refleksjonsfaktor for jorda
A = 4 * pi * r_jord**2   # Jordas kuleoverflate (m^2)
                                                                                
alfa = 0.6282  # Andel av atmosfærens stråling som kommer tilbake til jorda
beta = 0.8966  # Andel av jordas stråling som absorberes av atmosfæren
gamma = 0.3286 # Andel av solas innstråling (etter refleksjon) som absorberes av atmosfæren

S_sol = sigma * T_sol**4    # Solas strålingsstyrke ved soloverflaten (W/m^2)
S_0 = S_sol * r_sol**2/D**2 # Solas strålingsstyrke ved jorda (W/m^2)

P_inn = S_0 * (1 - albedo) * (pi * r_jord**2) # Solstrålingens effekt på jorda (W)

# --- Diff.likninger for temperaturendring ---

# Temperaturendring i atmosfæren
def T_a_der(T_j, T_a):
    return gamma * S_0 * (1 - albedo) + beta * (sigma * T_j**4) - (sigma * T_a**4)

# Temperaturendring i atmosfæren
def T_j_der(T_j, T_a):
    return (1 - gamma)*(P_inn/A) + alfa * (sigma * T_a**4) - (sigma * T_j**4)

eulers_metode(0, 0, T_j_der, T_a_der, utdyp=True)
```

### Modell 4

```python
from pylab import *

# --- Konstanter ---

sigma = 5.6703e-8       # Stefan-Boltzmanns konstant
T_sol = 5778            # Temperatur til sola (kelvin)
r_sol = 696_340_000     # Radius til sola (meter)
r_jord = 6_371e3        # Radius til jorda (meter)
D_min = 146_000_000_000 # Minste avstand mellom sola og jorda (meter)
D_max = 146_000_000_000 # Største avstand mellom sola og jorda (meter)
albedo = 0.3            # Refleksjonsfaktor for jorda
A = 4 * pi * r_jord**2  # Jordas kuleoverflate (m^2)

alfa = 0.5  # Andel av atmosfærens stråling som kommer tilbake til jorda.

norgefaktoren = 0.8 # Bestemmer hvor mye svakere S_0 er i Norge i forhold til ved ekvator
                    # Denne er helt vilkårlig valgt; den trenger bare å være mindre enn 1.

def D_sol_jord(dag):
    return -3e9 * cos(2*pi*dag/365) + 1.49e11


# --- Diff.likninger for temperaturendring ---

# Temperaturendring i atmosfæren
def T_a_der(T_j, T_a):
    return (sigma * T_j**4) - (sigma * T_a**4)

# Temperaturendring i atmosfæren
def T_j_der(T_j, T_a):
    return (P_inn/A) + alfa * (sigma * T_a**4) - (sigma * T_j**4)

# --- Beregning av temperatur for hver dag i året ---

aar = array(range(0, 365))

T_j_lst = []
T_a_lst = []

T_j = 0
T_a = 0


for dag in aar:
    # Beregne solinnstråling for dagen
    D = D_sol_jord(dag)
    S_sol = sigma * T_sol**4    # Solas strålingsstyrke ved soloverflaten (W/m^2)
    S_0 = (S_sol * r_sol**2/D**2)*norgefaktoren # Solas strålingsstyrke ved norge (W/m^2)

    P_inn = S_0 * (1 - albedo) * (pi * r_jord**2) # Solstrålingens effekt på jorda (W)
    
    # Bruke eulers metode -- Hvilken verdi stabiliserer temperaturen seg ved for denne dagen?
    T_j, T_a = eulers_metode(T_j, T_a, T_j_der, T_a_der, utdyp=False, returner=True)
    
    T_j_lst.append(T_j) # Logge jordas temperatur for denne dagen

# Plotte jordas temperatur i løpet av året

plot(aar, T_j_lst, label="Temperatur i Norge")

xlabel("Tid (dager etter 4. januar)")
ylabel("Temperatur (kelvin)")
legend()
show()

# Ekstra info
print(f"Største temperatur i løpet av et år: {round(max(T_j_lst) - 273.15, 1)}˚C")
print(f"Minste temperatur i løpet av et år: {round(min(T_j_lst) - 273.15, 1)}˚C")
print(f"Gjennomsnittstemperatur i løpet av et år: {round(mean(T_j_lst) - 273.15, 1)}˚C")
```

# Diskusjon


### Modell 1

Denne modellen er veldig forenklet. Det som hovedsaklig gjør denne modellen unøyaktig er at man ikke inkluderer en atmosfære. Da vil man ikke få med drivhuseffekten, som bidrar til å varme opp jorda, og resultatet blir ganske mye kaldere enn den målte verdien. Modellen gir -18˚C mens den målte gjennomsnittstemperaturen på jorda er rundt 15˚C. 

I denne modellen, samt modell 2 og 3, benytter vi oss av _gjennomsnittsavstanden_ til sola, selv om den i virkeligheten varierer i løpet av året. Man kan behandle resultatet da som et slags gjennomsnitt av temperaturene over et år, heller enn som at det skal stemme for noe bestemt tidspunkt.

### Modell 2

Her inkluderer vi en atmosfære i modellen for å få med drivhuseffektens virkning på jordas gjennomsnittstemperatur. Vi bruker diff.likninger for å modellere systemet som jorda og atmosfæren utgjør. På grafene kan man se at temperaturen til jorda og atmosfæren etter hvert stabiliserer seg.

Her blir resultatet betydelig varmere enn i modell 1, og nærmere den reelle verdien. Vi får 30˚C, så det er varmere enn den den målte verdien på 15˚C.

En av årsakene til at modellen er unøyaktig er at tallet vi bruker for solinnstrålingen kan være unøyaktig, siden vi ikke tar hensyn til at ikke alle steder på jorda får like mye solstråling som ved ekvator (eller egentlig bare det ene punktet på jorda som er nærmest sola) gjør.

En annen ting som gjør modellen unøyaktig er at atmosfæren er veldig forenklet. Vi antar at all solinnstråling går rett til jordoverflaten uten at noe blir absorbert av atmosfæren. I tillegg antar vi at atmosfæren absorberer 100% av jordas varmestråling, som den i realiteten ikke gjør. En tredje forenkling vi gjør er å si at atmosfæren i sum emitterer like mye varmestråling tilbake til jorda som ut til verdensrommet.

### Modell 3

I modell 3 utvider vi modell 2 for å gjøre atmosfære mer realistisk. Vi tar hensyn til at atmosfæren absorberer litt av solinnstrålingen, at den ikke absorberer 100% av jordas varmestråling og at den ikke utsråler like mye til jorda som til verdensrommet. Verdiene til koeffisientene som er brukt for å beskrive disse tingene i modellen er basert på faktiske verdier fra forskning rundt jordas energibudsjett.

Denne mer kompliserte atmosfæren gjør at temperaturen på jorda blir enda varmere enn i modell 3. Det er fordi antagelsene for modell 3 gjør at drivhuseffekten blir lavere enn i virkeligheten. Resultatet til denne modellen er en temperatur som er veldig mye høyere enn den målte verdien på 15˚C. Hovedgrunnen til dette kan igjen være at den gjennomsnittlige solinnstrålingen for hele jorda ikke er så høy som $S_0$, som jo gjelder kun for ekvator midt på dagen.

### Modell 4

I modell 4 utvider vi modell 2, denne gangen ikke for å ta hensyn til en mer realistisk atmosfære, men heller for å ta hensyn til den varierende avstanden mellom jorda og sola. Vi modellerer temperaturen til et punkt i Norge, midt på dagen, hver dag for et helt år. Den varierende avstanden vil gjøre at $S_0$, solas strålingsstyrke ved jorda blir høyere på når vi er nærmere sola og lavere når vi er lenger unna. Resultatet blir en temperatur på starten og slutten av året som er rundt 6˚C _varmere_ enn midt på året.

Noe som vi derimot ikke tar hensyn til i denne modellen er jordas helning. Helningen vil gjøre at breddegraden til Norge i praksis vil variere i løpet av året. (Her menes "breddegrad" relativt til jordas "topp- og bunnpunkt" i forhold til banen rundt sola, ikke relativt til polene, slik "breddegrad" egentlig er definert.) Denne variasjonen i breddegrad må påvirke solinnstrålingen i Norge over et helt år _mer_ enn variasjonen i avstand til sola. For steder på den nordlige halvkule, slik som Norge, må det være dette som gjør at det blir varmere midt på året enn på starten og slutten av året. Det er altså jordas helning som gir oss på den nordlige halvkule sommer midt på året og vinter på starten og slutten av året.

En unøyaktighet ved denne modellen er at funksjonen som brukes for å finne avstanden til sola for hver dag i året er en cosinusfunksjon. Det vil ikke nødvendigvis være tilfellet for jordas bane at avstanden varierer akkurat som en cosinusfunksjon; fysikken rundt planetbaner er ikke så lett. Å bruke en cosinusfunksjon kan på det beste kalles en tilnærming, og på det verste kanskje være utilgivelig feil.

En annen unøyaktighet for denne modellen er at akkurat hvor mye svakere solinnstrålingen er ved norge i forhold til ekvator er vilkårlig valgt. Men hensikten med denne modellen er ikke å finne riktige temperaturer, men å se på hvordan temperatur*endringen* ville sett ut i løpet av et år om jorda ikke hadde hatt noen helning.

# Konklusjon

Vi har altså sett på modeller for jordas temperatur av ulik kompleksitet, og selv om ingen av dem ga resultater som stemmer helt med den målte gjennomsnittstemperaturen på jorda, har de gitt oss innsikt i de forskjellige faktorene som påvirker temperaturen på jorda. I modell 1 så vi at jorda hadde vært betydelig kaldere uten atmosfære. Modell 2 viste oss hvor stor effekt selv en delvis forenklet atmosfære kunne ha for temperaturen. Modell 3 beskriver en mer kompleks atmosfære, men noen unøyaktigheter og antagelser gjør at heller ikke denne modellen gir særlig realistiske resultater. I modell 4 så vi hvordan temperaturen i et punkt på den nordlige halvkule hadde utviklet seg gjennom året om man så bort fra jordas helning, og modellen fortalte at årstidene ville vært mindre merkbare (pga. lavere temperaturforskjell) og at for den nordlige halvkule ville de vært snudd på hodet i forhold til virkeligheten.
