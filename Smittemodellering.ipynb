# Smittemodellering

```python
from pylab import *

data = loadtxt("influensa.txt", skiprows=2, delimiter=",")

obs_tid = data[:,0]
obs_verdier = data[:,1]

plot(obs_tid, obs_verdier)
show()
```

```python
I0 = 3
pop = 157_759
uker = 48

a = 4 / pop
b = 3.5

dt = 0.001

V = 0 # Andel vaksinerte
I = I0 # Smittede
J = I0 # Nye tilfeller
S = (pop - I) * (1 - V) # Mottakelige
R = 0 # Restituerte

I_list = [I0]
S_list = [S]
R_list = [R]
t_list = [0]
J_list = [0]

for t in arange(1, uker + 1, dt):
    I = I + a * S * I * dt - b * I * dt
    J = a * S * I
    S = S - a * S * I * dt
    R = R + b * I * dt

    R_list.append(R)
    S_list.append(S)
    I_list.append(I)
    t_list.append(t)
    J_list.append(J)

plot(t_list, S_list, label="S")
plot(t_list, I_list, label="I")
plot(t_list, R_list, label="R")

legend()
show()

plot(t_list, J_list, label="Nye tilfeller i modell")
plot(obs_tid, obs_verdier, label="Observert antall nye tilfeller")

legend()
show()
```
