# Numerisk derivasjon og integrasjon  Teori
___
### Derivasjon
Derivasjon av en funksjon $f(x)$ forteller hvor fort $f(x)$ endrer seg mellom punktet $x$ og $x + \Delta x$, der $\Delta x$ er et uendelig lite tall.

Dette kan skrives slik:

### $f'(x) = \lim\limits_{\Delta x \to 0}\frac{\Delta y}{\Delta x}$
ogs kjent som
### $f'(x) = \lim\limits_{\Delta x \to 0}\frac{f(x + \Delta x) - f(x)}{\Delta x}$

**Analytisk derivasjon** gjr man p en funksjon $f(x)$, og man fr et annet funksjonsuttrykk $f'(x)$ som gir den deriverte i en vilkrlig $x$.

Med **numerisk derivasjon**, derimot, regner man rett og slett $\frac{\Delta y}{\Delta x}$ for n bestemt $x$ og en *ganske liten* $\Delta x$ istedenfor en *uendelig liten* $\Delta x$.
Dette gir alts en tilnrmet verdi for $f'(x)$ for ett punkt. Tilnrmingen blir mer nyaktig for mindre $\Delta x$, fordi da kommer uttrykket nrmere den analytisk deriverte (som jo er eksakt). Jo nrmere $\Delta x$ kommer null, jo mindre blir feilen.
___
### Integrasjon
Integrasjon kan sees p som  finne arealet under grafen til en funksjon $f(x)$ mellom to $x$-verdier.

En numerisk metode for  gjre dette, er  dele opp arealet under grafen mellom to $x$-verdier $a$ og $b$ inn i $n$ stolper (rektangler). Summen av disse rektanglene vil vre tilnrmet lik det egentlige arealet under grafen. Jo flere rektangler, jo bedre blir tilnrmingen.

For  f et eksakt areal trenger man et uendelig antall rektangler ($n \to \infty$), mens med numerisk integrasjon har man bare et *veldig stort* antall rektangler (veldig stor $n$). Da fr man en tilnrming.
Jo strre $n$ er, jo nrmere kommer man den analytisk integrerte (hvor $n$ er uendelig stor), og jo bedre blir tilnrmingen.
